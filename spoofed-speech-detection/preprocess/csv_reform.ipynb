```python
import os
import pandas as pd
LABEL_FOLDER = '/data/chendan/protocal/CM_protocol/' #'ASV2015dataset/CM_protocol/' # Path to ground truth csv files
METHOD = "tacotron/"                                 #拓展方法，作为该文件夹总名称
WAV_FOLDER = "/data/chendan/wav/append/" + METHOD    #拓展方法./WAV文件夹,WAV文件夹下文件应全部为*.wav
EXT = ".wav"

subfolders = ['ELJ1']

train = pd.read_csv(LABEL_FOLDER + 'cm_train.trn', sep=' ', header=None, names=['folder','file','method','source'])
devel = pd.read_csv(LABEL_FOLDER + 'cm_develop.ndx', sep=' ', header=None, names=['folder','file','method','source'])
evalu = pd.read_csv(LABEL_FOLDER +'cm_evaluation.ndx', sep=' ', header=None, names=['folder','file','method','source'])

#只作为查看，运行不需要
##----------------------------------------------------------#
# if len(subfolders):                                       #
#     evalu = evalu[evalu.folder.isin(subfolders)]          #
# evalu.sort_values(['folder', 'file'], inplace=True)      #
##---------------------------------------------------------#

n_subfls = len(subfolders)
for folder in subfolders[0:n_subfls]:
    print(folder)
    new_folder = "".join((WAV_FOLDER,folder,'/'))
    # print(folder)
    f_list = os.listdir(new_folder)
    # print(f_list)
    for f_name in f_list:
        sf_name = f_name[:-len(EXT)]
        if sf_name not in evalu.file.values:
            evalu = evalu.append({'folder': folder, 'file': sf_name, 'method': METHOD[:-1], 'source': 'spoof'},ignore_index=True)
```

```python
evalu.to_csv(LABEL_FOLDER + 'cm_evaluation_0103.ndx', sep=" ", header=0)
```
